```python
from rdkit import Chem
from rdkit.Chem import BRICS

mol = Chem.MolFromSmiles("Cc1cc(Br)ccc1N=C(S)N1CCOCC1")#("O=C(CSc1c[nH]c2ccccc12)N(Cc1ccccc1)Cc1ccccc1") #Chem.MolFromSmiles("O=C(O[C@@H]2Cc3c(O[C@@H]2c1ccc(O)c(O)c1)cc(O)cc3O)c4cc(O)c(O)c(O)c4") #  Chem.MolFromSmiles("c1ccc(cc1)Nc2ccccc2")
mol.UpdatePropertyCache(strict=True)

brics_bonds = list(BRICS.FindBRICSBonds(mol))

broken_bonds = [(bond[0][0], bond[0][1]) for bond in brics_bonds]
print("Разорванные связи (i, j):", broken_bonds)
```
